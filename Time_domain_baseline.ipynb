```python
import pandas as pd
import numpy as np
import glob
import os
from scipy.stats import skew, kurtosis

# Path to your extracted partition2_instances folder
partition_path = '/content/partition4/NF'  # <-- UPDATE if needed

# Columns you are interested in
columns_of_interest = ['TOTUSJH', 'TOTPOT', 'TOTUSJZ']

# Feature extraction function
def extract_time_features(file_path):
    df = pd.read_csv(file_path, delimiter='\t', usecols=columns_of_interest)

    features = []
    for col in df.columns:
        signal = df[col].dropna().values
        features.append(np.mean(signal))
        features.append(np.std(signal))
        features.append(np.min(signal))
        features.append(np.max(signal))
        features.append(np.percentile(signal, 25))
        features.append(np.percentile(signal, 50))
        features.append(np.percentile(signal, 75))
        features.append(skew(signal))
        features.append(kurtosis(signal))
    return features

# Process all CSVs in F folder
X = []
file_names = []

csv_files = glob.glob(os.path.join(partition_path, '*.csv'))

for file_path in csv_files:
    features = extract_time_features(file_path)
    file_name = os.path.basename(file_path)
    subcategory = file_name[:2]  # Raw subcategory (e.g., 'F1', 'C3')
    features.append(subcategory)
    X.append(features)
    file_names.append(file_name)

# Convert to NumPy array
X = np.array(X)

# Define column names
columns = []
for col in columns_of_interest:
    columns += [
        f"{col}_mean", f"{col}_std", f"{col}_min", f"{col}_max",
        f"{col}_q25", f"{col}_q50", f"{col}_q75", f"{col}_skew", f"{col}_kurtosis"
    ]
columns.append('subcategory')  # Last column is subcategory

# Create DataFrame
df_features_F = pd.DataFrame(X, columns=columns)

# Clean subcategory to remove digits (e.g., 'F1' -> 'F')
df_features_F['subcategory'] = df_features_F['subcategory'].str.extract(r'([A-Za-z]+)')

# Add file names and label
df_features_F['file_name'] = file_names
df_features_F['label'] = 0  #  Set label = 1 for this class

# Save the dataset
df_features_F.to_csv('time_domain_features_partition4_NF.csv', index=False)

print(" Feature extraction for F folder complete!")
print(" Saved as: time_domain_features_partition4_NF.csv")
```

##SAMPLING

```python
def fetch_samples(df_comb):
    fl_records = df_comb[df_comb['label'] == 1]  # FL = 1
    fl_count = len(fl_records)

    num_c = int((3 / 6) * fl_count)  # 50% from C
    num_b = int((2 / 6) * fl_count)  # 33.33% from B
    num_fq = fl_count - (num_c + num_b)  # Remaining from FQ

    # Sample without replacement and handle cases where available data is insufficient
    c_samples = df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'C')].sample(
        n=min(num_c, len(df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'C')])),
        random_state=42, replace=False
    )

    b_samples = df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'B')].sample(
        n=min(num_b, len(df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'B')])),
        random_state=42, replace=False
    )

    remaining_needed = fl_count - (len(c_samples) + len(b_samples))

    fq_samples = df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'FQ')].sample(
        n=min(remaining_needed, len(df_comb[(df_comb['label'] == 0) & (df_comb['subcategory'] == 'FQ')])),
        random_state=42, replace=False
    )

    balanced_sample = pd.concat([fl_records, c_samples, b_samples, fq_samples])
    balanced_sample = balanced_sample.sample(frac=1, random_state=42).reset_index(drop=True)

    return balanced_sample
```

```python
df_train = pd.concat([fetch_samples(df1), fetch_samples(df2), fetch_samples(df3)], axis=0)
```

```python
from sklearn.metrics import (
    accuracy_score, classification_report, precision_score, recall_score,
    f1_score, jaccard_score, confusion_matrix, balanced_accuracy_score
)
from imblearn.metrics import geometric_mean_score

# Define the evaluation functions
def evaluate_metrics(y_true, y_pred):
    pos_lbl = 0 if y_true.dtype in ['int64', 'int32'] else 'FL'
    metrics = {}
    metrics['balanced_accuracy'] = balanced_accuracy_score(y_true, y_pred)
    metrics['geometric_mean'] = geometric_mean_score(y_true, y_pred)
    metrics['f1_macro'] = f1_score(y_true, y_pred, average='macro')
    metrics['f1_weighted'] = f1_score(y_true, y_pred, average='weighted')
    metrics['accuracy'] = accuracy_score(y_true, y_pred)
    metrics['precision_macro'] = precision_score(y_true, y_pred, average='macro')
    metrics['precision_weighted'] = precision_score(y_true, y_pred, average='weighted')
    metrics['recall_macro'] = recall_score(y_true, y_pred, average='macro')
    metrics['recall_weighted'] = recall_score(y_true, y_pred, average='weighted')
    metrics['jaccard_macro'] = jaccard_score(y_true, y_pred, average='macro')
    metrics['jaccard_weighted'] = jaccard_score(y_true, y_pred, average='weighted')
    metrics['confusion_matrix'] = confusion_matrix(y_true, y_pred)

    # Ensure confusion matrix is 2x2
    if metrics['confusion_matrix'].shape == (2, 2):
        tp, fn, fp, tn = metrics['confusion_matrix'].ravel()
        tss = (tp / (tp + fn)) - (fp / (fp + tn))
        hss = (2 * (tp * tn - fp * fn)) / ((tp + fn) * (fn + tn) + (tp + fp) * (fp + tn))
        metrics['tss'] = tss
        metrics['hss'] = hss
    else:
        metrics['tss'] = None
        metrics['hss'] = None

    metrics['classification_report'] = classification_report(y_true, y_pred)
    return metrics

def print_metrics(metrics, title):
    print(title, 'Results')
    print('=' * 40)
    for k, v in metrics.items():
        if k in ('confusion_matrix', 'classification_report'):
            print(f'{k:<20}')
            print(v, '\n')
            continue
        print(f'{k:<20}: {v:.5f}' if v is not None else f'{k:<20}: N/A')

# === Make sure features match
df_test_cleaned = df_test[X.columns]  # Ensure same columns used as in training
y_test_true = df_test['label']

# === Predict using your trained RF model
y_test_pred = rf_clf.predict(df_test_cleaned)
y_test_proba = rf_clf.predict_proba(df_test_cleaned)

# === Evaluate
rf_metrics = evaluate_metrics(y_test_true, y_test_pred)

# === Save and print
metrics_dict = {}
metrics_dict['random_forest'] = rf_metrics
print_metrics(rf_metrics, title='Random Forest Classifier')
```
